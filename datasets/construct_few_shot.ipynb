```python
# select few-shot images from CIFAR-100 dataset
import os
import shutil

def select_and_copy_images(src_folder, dst_folder):
    # 创建目标文件夹
    if not os.path.exists(dst_folder):
        os.makedirs(dst_folder)

    # 遍历源文件夹中的所有子文件夹
    for subdir in os.listdir(src_folder):
        src_subdir_path = os.path.join(src_folder, subdir)
        
        # 检查是否为文件夹
        if os.path.isdir(src_subdir_path):
            dst_subdir_path = os.path.join(dst_folder, subdir)
            
            # 在目标文件夹中创建对应的子文件夹
            if not os.path.exists(dst_subdir_path):
                os.makedirs(dst_subdir_path)

            # 获取所有图片文件
            images = [f for f in os.listdir(src_subdir_path) if os.path.isfile(os.path.join(src_subdir_path, f))]
            
            # 按文件名排序
            images.sort()

            # 选择前16张图片
            selected_images = images[:16]

            # 复制选定的图片到目标文件夹
            for img in selected_images:
                src_img_path = os.path.join(src_subdir_path, img)
                dst_img_path = os.path.join(dst_subdir_path, img)
                shutil.copy2(src_img_path, dst_img_path)

# 使用示例
src_folder = 'data/cifar100_images'  # 源文件夹路径
dst_folder = 'data/cifar100_few_shot'  # 目标文件夹路径

select_and_copy_images(src_folder, dst_folder)
```

```python
# count few-shot images
import os
import shutil

def select_and_copy_images(src_folder, dst_folder, num=16):
    for root, dirs, files in os.walk(src_folder):
        for subdir in dirs:
            src_subdir_path = os.path.join(root, subdir)
            relative_path = os.path.relpath(src_subdir_path, src_folder)
            dst_subdir_path = os.path.join(dst_folder, relative_path)

            # 在目标文件夹中创建相对应的子文件夹
            if not os.path.exists(dst_subdir_path):
                os.makedirs(dst_subdir_path)

        # 获取当前文件夹中的所有图片文件
        images = [f for f in files if os.path.isfile(os.path.join(root, f))]
        # 按文件名排序
        images.sort()
        # 选择前num张图片
        selected_images = images[:num]

        # 复制选定的图片到目标文件夹
        for img in selected_images:
            src_img_path = os.path.join(root, img)
            dst_img_path = os.path.join(dst_folder, os.path.relpath(root, src_folder), img)
            shutil.copy2(src_img_path, dst_img_path)

# 使用示例
num = 1  # 每个类别选择的图片数量            
src_folder = 'data/cifar100_images'  # 源文件夹路径
dst_folder = 'data/cifar100_few_shot_' + str(num) # 目标文件夹路径

select_and_copy_images(src_folder, dst_folder, num=num)
```
